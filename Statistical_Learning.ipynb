4.Write a function with name â€œGoal_diff_countâ€ which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. (7 points)

```python
def Goal_diff_count(data):
    data['GoalsFor'] = data['GoalsFor'].apply(pd.to_numeric)
    data['GoalsAgainst'] = data['GoalsAgainst'].apply(pd.to_numeric)
    data.insert(10,'Goal Difference',data['GoalsFor'] - data['GoalsAgainst'])    
    return data,data[data['Goal Difference']==data['Goal Difference'].max()],data[data['Goal Difference']==data['Goal Difference'].min()]
```

```python
goalDifference,maxdiff,mindiff = Goal_diff_count(df.copy())  
```

Goal Difference by team 

```python
goalDifference[['Team','GoalsFor','GoalsAgainst','Goal Difference']]
```

The team which has maximum goal difference

```python
maxdiff[['Team','GoalsFor','GoalsAgainst','Goal Difference']]
```

The team which has minimum goal difference

```python
mindiff[['Team','GoalsFor','GoalsAgainst','Goal Difference']]
```
